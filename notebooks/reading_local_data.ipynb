This notebook is a short walkthrough of how to read and begin to analyse the data included in this repository.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
SAMPLE_RATE = 25000
```

Let's being by getting the data and converting it from CSV to a Pandas DataFrame so it's easier to visualise. It's probably simpler to manipulate in NumPy format but it's easier to visualise using Pandas.

You can essentially read any of the CSV files ('grp_data', 'spontaneous_data' or 'ttx_data') but we'll just look at grp_data for now.

If you want to know the details of this data, be sure to check out the `data.md` file in the `data` folder.

```python
# load csv into pandas dataframe
df = pd.read_csv("../data/grp_data.csv")

print(df.shape)

df.head()
```

The dataframe has a shape of (100000, 60) representing 100000 datapoints in 60 recording channels. As the recording is taken at 25kHz, this is equivelent to 4 seconds of data.

Let's isolate one of the channels at see what it looks like.

```python
# choose a channel
channel = 10

# convert this column to numpy array
data = df[f"{channel}"].to_numpy()

# create a time array for the x axis
time = np.arange(0, len(data)) / SAMPLE_RATE

# create time series plot
plt.figure(figsize=(18, 4))
plt.plot(time, data, label='Raw Data', c="black", lw=0.5)
plt.ylabel('Voltage')
plt.xlabel('Time (s)')
plt.xlim(0, max(time))
```
